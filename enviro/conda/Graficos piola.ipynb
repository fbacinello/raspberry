```python
df2 = pd.read_csv(filepath)
df2.head()
```

```python
df2["temp"] = pd.to_numeric(df2["temp"], downcast="float")
df2['time'] = pd.to_datetime(df2['time'], format='YYYY-MM-DD[*HH[:MM[:SS[.fff[fff]]]]')
print(df2.dtypes)
```

```python

df2.sort_values('time').iplot(
    x='time',
    y='temp',
    xTitle='Read Time',
    yTitle='Reading Percent',
    title='Reading Percent vs Reading Time',
    mode='markers+lines',
    bestfit=True, bestfit_colors=['blue'])
```
