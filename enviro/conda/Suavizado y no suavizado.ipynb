```python
import pandas as pd
import matplotlib.pyplot as plt

filepath = '../data/enviro_suav08_08_2020.csv'
df_suav = pd.read_csv(filepath)
df_suav.head()

filepath = '../data/enviro_not_suav08_08_2020.csv'
df_no_suav = pd.read_csv(filepath)
df_no_suav.head()
```

```python
y_data = df_suav['humi']
y_label = 'Humedad'
x_data = df_suav['time']
x_label = 'Tiempo'
title = "Humedad en el tiempo suavizada"

fig, ax = plt.subplots()
ax.scatter(x_data, y_data, s = 30, color = '#539caf', alpha = 0.75)

ax.set_title(title)
ax.set_xlabel(x_label)
ax.set_ylabel(y_label)
fig.autofmt_xdate()
```
