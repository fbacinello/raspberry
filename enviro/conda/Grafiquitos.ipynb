```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filepath = '../data/enviro.csv'
#df = pd.read_csv(filepath, index_col = ['time'])
df = pd.read_csv(filepath)
```

```python
y_data = df['humi']
y_label = 'Humedad'
x_data = df['time']
x_label = 'Tiempo'
title = "Humedad en el tiempo"
```

```python
fig, ax = plt.subplots()
ax.scatter(x_data, y_data, s = 30, color = '#539caf', alpha = 0.75)

ax.set_title(title)
#ax.set_xlabel(x_label)
ax.set_ylabel(y_label)
fig.autofmt_xdate()
```
